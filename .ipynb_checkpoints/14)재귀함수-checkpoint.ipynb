# 재귀호출
- 함수 안에서 함수 자기 자신을 호출하는 방식
- 알고리즘을 구현할 때 반복문으로 구현한 코드보다 재귀호출로 구현한 코드가 더 직관적이고 이해하기 쉬운 경우가 종종 있음

## 재귀호출 사용

```python
def hello():
    print("Hello, World!")
    hello()
```

- 재귀호출이 종료되지 않으면 최대 재귀 깊이를 초과할 때 에러 발생
- 파이썬의 최대 재귀 깊이(maximum recursion depth) : 1000

```python
hello()
```

## 재귀호출에 종료 조건 만들기

```python
def hello(count):
    if count == 0: # 종료 조건, count가 0일 때 종료
        return
    print("Hello World!", count)
    count -= 1 # count를 1감소 시킨뒤 다시 hello에 넣음
    hello(count)
```

```python
hello(5)
```

## 팩토리얼 구하기
- 팩토리얼
    - 1부터 n까지 양의 정수를 차례로 곱한 값
    - !기호로 표기함
    - 예) 5! = 5 * 4 * 3 * 2 * 1 = 120

- 팩토리얼 구하는 방법
    - 반복문
    - 재귀함수

## 반복문으로 팩토리얼 구하기

```python
def factorial(n):
    output = 1
    
    # 반복문으로 숫자를 더하기
    for i in range(1, n + 1):
        output *= i
        
    return output
```

```python
factorial(3)
```

```python
factorial(4)
```

```python
factorial(5)
```

## 재귀함수로 팩토리얼 구하기

```python
def factorial(n):
    if n == 1: # n 이 1일때, 1을 반환하고 재귀호출 종료
        return 1
    
    return n * factorial(n - 1) # n과 factorial 함수에 n - 1을 넣어서 반환된 값을 곱하기
```

```python
factorial(5)
```

- factorial 함수 연산 과정
- 5 * factorial(4)
    - 4 * factorial(3)
        - 3 * factorial(2)
            - 2 * factorial(1)
                - return 1
            - 2 * 1
        - 3 * 2
    - 4 * 6
- 5 * 24

# 재귀함수의 문제점
- 상황에 따라서 같은 것을 기하급수적으로 많이 반복하게됨
    - 이 문제를 해결하기 위해서 메모화(memorization)를 사용할 수 있음

```python
# 피보나치 수열 : 0과 1로 시작해서 다음번 수는 바로 앞의 두 피보나치 수의 합
def fibonacci(n):
    if n == 1 or n == 2:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
fibonacci(10)
```

```python
fibonacci(30)
```

```python
%%time
fibonacci(35)
```

```python
counter = 0

def fibonacci(n):
    print(f"fibonacci({n})를 계산중")
    global counter
    counter += 1
    
    # 피보나치 구하기
    if n == 1 or n == 2:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
fibonacci(10)
```

```python
print(counter)
```

- 재귀함수는 한 번 구했던 값이라도 처음부터 다시 계산해야 하기 때문에 계산 횟수가 기하급수적으로 늘어난다.

## 메모화
- 재귀함수의 문제는 같은 값을 구하는 연산을 여러 번 반복하고 있기 때문에 발생
    - 같은 값을 한 번만 계산하도록 메모화를 사용

```python
memo_dict = {1 : 1, 2 : 1}

def fibonacci(n):
    if n in memo_dict:
        # 메모가 되어 있으면 메모된 값을 리턴
        return memo_dict[n]
    else:
        # 메모가 되어있지 않으면 값을 구함
        output = fibonacci(n - 1) + fibonacci(n - 2)
        memo_dict[n] = output
        return output
```

```python
%%time
fibonacci(50)
```

- 딕셔너리를 사용해서 한 번 계산한 값을 저장하였음
    - 메모(memo)
    - 딕셔너리에 값이 저장되어 있으면 처리를 수행하지않고 곧바로 메모된 값을 반환하면서 코드의 속도가 빨라짐

# 리스트 평탄화하는 재귀함수
- 리스트 평탄화 : 중첩된 리스트가 있을 때 중첩을 모두 제거하고 풀어서 1차원 리스트로 만드는 것

```python
def flatten(data):
    output = []
    
    for item in data:
        # 리스트가 아니라면 output리스트에 자료를 넣고
        # 리스트라면 리스트에 있는 요소들을 하나하나 output에 추가
        if type(item) == list:
            output.extend(item)
        else:
            output.append(item)
            
    return output
```

```python
example = [[1, 2, 3,], [4, 5, 6], 7, [8, 9]]
flatten(example)
```

```python
# 3차원이 되면 문제 발생
example = [[1, 2, 3,], [4, [5, 6]], 7, [8, 9]]
flatten(example)
```

```python
# 재귀함수 사용
def flatten(data):
    output = []
    
    for item in data:
        if type(item) == list:
            output.extend(flatten(item))
            # output += flatten(itme)
        else:
            output.append(item)
    return output
```

```python
example = [[1, 2, 3], [4, 5, 6], 7, [8, 9]]
flatten(example)
```

```python
example = [[1, 2, 3], [4, [5, 6]], 7, [8, 9]]
flatten(example)
```

```python
example = [[1, 2, 3], [4, [5, [6, 7]]], [8, 9]]
flatten(example)
```

```python
# 연습문제
# 재귀호출로 회문 판별하기
# 어디서 재귀를 해야될지 잘 모르겠음
def is_palindrome(word):
    count = 0
    rev_count = len(word)
    is_true = False
    for i in range(len(word)):
        if word[count] != word[rev_count - 1]:
            is_ture = False
            break
        else:
            is_true = True
        count += 1
        rev_count -= 1
    
    return is_true
```

```python
is_palindrome("hello")
```

```python
is_palindrome("level")
```

```python
# 강사님 풀이
def is_palindrome(word):
    if len(word) < 2:
        return True
    
    if word[0] != word[-1]:
        return False
    
    return is_palindrome(word[1:-1])
```

```python
is_palindrome("hello")
```

```python
is_palindrome("level")
```
